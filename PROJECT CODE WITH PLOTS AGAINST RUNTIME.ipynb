```python
#Rewards of all states, R is for the goal state, Rr is for the negative terminal state and r is for all other states
R,Rr,r=1,-1,0
```

```python
States=[(1,1),(1,2),(1,3),(2,3),(2,1),(3,1),(3,2),(3,3),(4,1)]
```

```python
Actions=['L','R','U','D']# The list of all possible actions
```

```python
alpha=0.001 #learning rate
gamma=0.9  #discounting factor
epsilon=0.9 #exploration rate
epsilon_min=0.01
```

```python
def E_greedy1():
    ActionValues=[round(random.uniform(0.1,0.3),4) for i in range(len(States)*4)]
    #N=50000
    start = time.time()
    for i in range(N):
        S=random.choice(States) #initialising the statrting state for each episode

       # print(S,A)
        while S!=(4,2) and S!=(4,3):

      #EXPLORATION ACTION
            if random.uniform(0,1)>1-epsilon:# exploration condition
                #EXPLORATION ACTION
            #if random.uniform(0,1)>1-epsilon:# exploration condition
                A=random.choice(Actions)     #selecting a random action 
    #STATE (1,1)##########################################################################

                if S==(1,1):
                    if A=='U':
                        ActionValues[0]+=alpha*(r+gamma*max(ActionValues[4:8])-ActionValues[0]) #Q-learning update rule
                        S=(1,2)

                    elif A=='D':
                        ActionValues[1]+=alpha*(r+gamma*max(ActionValues[0:4])-ActionValues[1])

                        S=(1,1)

                    elif A=='L':
                        ActionValues[2]+=alpha*(r+gamma*max(ActionValues[0:4])-ActionValues[2])
                        S=(1,1)

                    elif A=='R':
                        ActionValues[3]+=alpha*(r+gamma*max(ActionValues[16:20])-ActionValues[3])
                        S=(2,1)

                #STATE (1,2)###############################################################

                elif S==(1,2):
                    if A=='U':
                        ActionValues[4]+=alpha*(r+gamma*max(ActionValues[8:12])-ActionValues[4])
                        S=(1,3)

                    elif A=='D':
                        ActionValues[5]+=alpha*(r+gamma*max(ActionValues[0:4])-ActionValues[5])
                        S=(1,1)

                    elif A=='L':
                        ActionValues[6]+=alpha*(r+gamma*max(ActionValues[4:8])-ActionValues[6])
                        S=(1,2)

                    elif A=='R':
                        ActionValues[7]+=alpha*(r+gamma*max(ActionValues[4:8])-ActionValues[7])
                        S=(2,1)

                #STATE (1,3)#################################################################

                elif S==(1,3):
                    if A=='U':
                        ActionValues[8]+=alpha*(r+gamma*max(ActionValues[8:12])-ActionValues[8])
                        S=(1,3)

                    elif A=='D':
                        ActionValues[9]+=alpha*(r+gamma*max(ActionValues[4:8])-ActionValues[9])
                        S=(1,2)

                    elif A=='L':
                        ActionValues[10]+=alpha*(r+gamma*max(ActionValues[8:12])-ActionValues[10])
                        S=(1,3)

                    elif A=='R':
                        ActionValues[11]+=alpha*(r+gamma*max(ActionValues[12:16])-ActionValues[11])
                        S=(2,3)

                #STATE (2,1)##################################################################
                elif S==(2,1):
                    if A=='U':
                        ActionValues[16]+=alpha*(r+gamma*max(ActionValues[16:20])-ActionValues[16])
                        S=(2,1)

                    elif A=='D':
                        ActionValues[17]+=alpha*(r+gamma*max(ActionValues[16:20])-ActionValues[17])
                        S=(2,1)

                    elif A=='L':
                        ActionValues[18]+=alpha*(r+gamma*max(ActionValues[0:4])-ActionValues[18])
                        S=(1,1)

                    elif A=='R':
                        ActionValues[19]+=alpha*(r+gamma*max(ActionValues[20:24])-ActionValues[19])
                        S=(3,1)

                #STATE (2,3)#####################################################################
                elif S==(2,3):
                    if A=='U':
                        ActionValues[12]+=alpha*(r+gamma*max(ActionValues[12:16])-ActionValues[12])
                        S=(2,3)

                    elif A=='D':
                        ActionValues[13]+=alpha*(r+gamma*max(ActionValues[12:16])-ActionValues[13])
                        S=(2,3)

                    elif A=='L':
                        ActionValues[14]+=alpha*(r+gamma*max(ActionValues[8:12])-ActionValues[14])
                        S=(1,3)

                    elif A=='R':
                        ActionValues[15]+=alpha*(r+gamma*max(ActionValues[28:32])-ActionValues[15])
                        S=(3,3)

                #STATE (3,1)#######################################################################


                elif S==(3,1):
                    if A=='U':
                        ActionValues[20]+=alpha*(r+gamma*max(ActionValues[24:28])-ActionValues[20])
                        S=(3,2)

                    elif A=='D':
                        ActionValues[21]+=alpha*(r+gamma*max(ActionValues[20:24])-ActionValues[21])
                        S=(2,3)

                    elif A=='L':
                        ActionValues[22]+=alpha*(r+gamma*max(ActionValues[16:20])-ActionValues[22])
                        S=(2,1)

                    elif A=='R':
                        ActionValues[23]+=alpha*(r+gamma*max(ActionValues[32:])-ActionValues[23])
                        S=(4,1)

                #STATE (3,2)#####################################################################


                elif S==(3,2):
                    if A=='U':
                        ActionValues[24]+=alpha*(r+gamma*max(ActionValues[28:32])-ActionValues[24])
                        S=(3,2)

                    elif A=='D':
                        ActionValues[25]+=alpha*(r+gamma*max(ActionValues[20:24])-ActionValues[25])

                        S=(2,1)

                    elif A=='L':
                        ActionValues[26]+=alpha*(r+gamma*max(ActionValues[24:28])-ActionValues[26])

                        S=(3,2)

                    elif A=='R':
                        ActionValues[27]+=alpha*(Rr+gamma*max(0,0,0,0)-ActionValues[27])

                        S=(4,2)

                #STATE (3,3)########################################################################


                elif S==(3,3):
                    if A=='U':
                        ActionValues[28]+=alpha*(r+gamma*max(ActionValues[28:32])-ActionValues[28])

                        S=(3,3)

                    elif A=='D':
                        ActionValues[29]+=alpha*(r+gamma*max(ActionValues[24:28])-ActionValues[29])

                        S=(3,2)

                    elif A=='L':
                        ActionValues[30]+=alpha*(r+gamma*max(ActionValues[12:16])-ActionValues[30])

                        S=(2,3)

                    elif A=='R':
                        ActionValues[31]+=alpha*(R+gamma*max(0,0,0,0)-ActionValues[31])
                        S=(4,3)

                #STATE (4,1)#######################################################################
                elif S==(4,1):
                    if A=='U':
                        ActionValues[32]+=alpha*(Rr+gamma*max(0,0,0,0)-ActionValues[32])
                        S=(4,2)

                    elif A=='D':
                        ActionValues[33]+=alpha*(r+gamma*max(ActionValues[32:])-ActionValues[33])
                        S=(4,1)

                    elif A=='L':
                        ActionValues[34]+=alpha*(r+gamma*max(ActionValues[20:24])-ActionValues[34])

                        S=(3,1)

                    elif A=='R':
                        ActionValues[35]+=alpha*(r+gamma*max(ActionValues[32:])-ActionValues[35])

                        S=(4,1)

                        
                        
            if random.uniform(0,1)<1-epsilon: #condition for exploitation

                #STATE(1,1)
                if S==(1,1):
                    c=ActionValues.index(max(ActionValues[:4])) #selecting the maximum action in state (1,1)
                    if c==0:#up
                        ActionValues[c]+=alpha*(r+gamma*max(ActionValues[4:8])-ActionValues[c])
                        S=(1,2)

                    if c==1:#down
                        ActionValues[c]+=alpha*(r+gamma*max(ActionValues[0:4])-ActionValues[c])
                        S=(1,1)

                    if c==2:#left
                        ActionValues[c]+=alpha*(r+gamma*max(ActionValues[0:4])-ActionValues[c])
                        S=(1,1)

                    if c==3:#right
                        ActionValues[c]+=alpha*(r+gamma*max(ActionValues[16:20])-ActionValues[c])
                        S=(1,1)

                #STATE(1,2)
                if S==(1,2):
                    c=ActionValues.index(max(ActionValues[4:8]))
                    if c==4:#up
                        ActionValues[c]+=alpha*(r+gamma*max(ActionValues[8:12])-ActionValues[c])
                        S=(1,3)

                    if c==5:#down
                        ActionValues[c]+=alpha*(r+gamma*max(ActionValues[0:4])-ActionValues[c])
                        S=(1,1)

                    if c==6:#left
                        ActionValues[c]+=alpha*(r+gamma*max(ActionValues[4:8])-ActionValues[c])
                        S=(1,2)

                    if c==7:#right
                        ActionValues[c]+=alpha*(r+gamma*max(ActionValues[4:8])-ActionValues[c])
                        S=(1,2)

                #STATE(1,3)
                if S==(1,3):
                    c=ActionValues.index(max(ActionValues[8:12]))
                    if c==8:#up
                        ActionValues[c]+=alpha*(r+gamma*max(ActionValues[8:12])-ActionValues[c])
                        S=(1,3)

                    if c==5:#down
                        ActionValues[c]+=alpha*(r+gamma*max(ActionValues[4:8])-ActionValues[c])
                        S=(1,2)

                    if c==6:#left
                        ActionValues[c]+=alpha*(r+gamma*max(ActionValues[8:12])-ActionValues[c])
                        S=(1,3)

                    if c==7:#right
                        ActionValues[c]+=alpha*(r+gamma*max(ActionValues[12:16])-ActionValues[c])
                        S=(2,3)

                #STATE(2,3)
                if S==(2,3):
                    c=ActionValues.index(max(ActionValues[12:16]))
                    if c==12:#up
                        ActionValues[c]+=alpha*(r+gamma*max(ActionValues[12:16])-ActionValues[c])
                        S=(2,3)

                    if c==13:#down
                        ActionValues[c]+=alpha*(r+gamma*max(ActionValues[12:16])-ActionValues[c])
                        S=(2,3)

                    if c==14:#left
                        ActionValues[c]+=alpha*(r+gamma*max(ActionValues[8:12])-ActionValues[c])
                        S=(1,3)

                    if c==15:#right
                        ActionValues[c]+=alpha*(r+gamma*max(ActionValues[28:32])-ActionValues[c])
                        S=(3,3)

                #STATE(2,1)
                if S==(2,1):
                    c=ActionValues.index(max(ActionValues[16:20]))
                    if c==16:#up
                        ActionValues[c]+=alpha*(r+gamma*max(ActionValues[16:20])-ActionValues[c])
                        S=(2,1)

                    if c==17:#down
                        ActionValues[c]+=alpha*(r+gamma*max(ActionValues[16:20])-ActionValues[c])
                        S=(2,1)

                    if c==18:#left
                        ActionValues[c]+=alpha*(r+gamma*max(ActionValues[:4])-ActionValues[c])
                        S=(1,1)

                    if c==19:#right
                        ActionValues[c]+=alpha*(r+gamma*max(ActionValues[20:24])-ActionValues[c])
                        S=(3,1)


                #STATE(3,1)
                if S==(3,1):
                    c=ActionValues.index(max(ActionValues[20:24]))
                    if c==20:#up
                        ActionValues[c]+=alpha*(r+gamma*max(ActionValues[24:28])-ActionValues[c])
                        S=(3,2)

                    if c==21:#down
                        ActionValues[c]+=alpha*(r+gamma*max(ActionValues[20:24])-ActionValues[c])
                        S=(3,1)

                    if c==22:#left
                        ActionValues[c]+=alpha*(r+gamma*max(ActionValues[16:20])-ActionValues[c])
                        S=(2,1)

                    if c==22:#right
                        ActionValues[c]+=alpha*(r+gamma*max(ActionValues[20:24])-ActionValues[c])
                        S=(4,1)


                #STATE(3,2)
                if S==(3,2):
                    c=ActionValues.index(max(ActionValues[24:28]))
                    if c==24:#up
                        ActionValues[c]+=alpha*(r+gamma*max(ActionValues[28:32])-ActionValues[c])
                        S=(3,3)

                    if c==25:#down
                        ActionValues[c]+=alpha*(r+gamma*max(ActionValues[20:24])-ActionValues[c])
                        S=(3,1)

                    if c==26:#left
                        ActionValues[c]+=alpha*(r+gamma*max(ActionValues[24:28])-ActionValues[c])
                        S=(3,2)

                    if c==27:#right
                        # the q values of terminal states are all 0's
                        ActionValues[c]+=alpha*(Rr+gamma*max(0,0,0,0)-ActionValues[c])
                        S=(4,2)

                #STATE(3,3)
                if S==(3,3):
                    c=ActionValues.index(max(ActionValues[28:32]))
                    if c==28:#up
                        ActionValues[c]+=alpha*(r+gamma*max(ActionValues[28:32])-ActionValues[c])
                        S=(3,3)

                    if c==29:#down
                        ActionValues[c]+=alpha*(r+gamma*max(ActionValues[24:28])-ActionValues[c])
                        S=(3,2)

                    if c==30:#left
                        ActionValues[c]+=alpha*(r+gamma*max(ActionValues[12:16])-ActionValues[c])
                        S=(2,3)

                    if c==31:#right
                        ActionValues[c]+=alpha*(R+gamma*max(0,0,0,0)-ActionValues[c])
                        S=(4,1)


                #STATE(4,1)
                if S==(4,1):
                    c=ActionValues.index(max(ActionValues[32:]))
                    if c==32:#up
                        ActionValues[c]+=alpha*(Rr+gamma*max(0,0,0,0)-ActionValues[c])
                        S=(4,2)

                    if c==33:#down
                        ActionValues[c]+=alpha*(r+gamma*max(ActionValues[32:])-ActionValues[c])
                        S=(4,1)

                    if c==34:#left
                        ActionValues[c]+=alpha*(r+gamma*max(ActionValues[20:24])-ActionValues[c])
                        S=(3,1)

                    if c==35:#right
                        ActionValues[c]+=alpha*(r+gamma*max(ActionValues[32:])-ActionValues[c])
                        S=(4,1)
        
                        
                

    time.sleep(1)
    end = time.time()
    print(f"Runtime of the program is {end - start}")
    return ActionValues
```

```python
def E_greedy2():
    #N=50000
    ActionValues=[round(random.uniform(0.1,0.3),4) for i in range(len(States)*4)]
    sum_mse=[]
    #ss.append(sum_mse)
    start = time.time()
    for i in range(N):
        S=random.choice(States) #initialising the statrting state for each episode

       # print(S,A)
        while S!=(4,2) and S!=(4,3):
                #finding the sum squared error to come up with 1 plot from 36 different Q-values
            sum_mse.append(((ActionValues[0]-t[0])**2+(ActionValues[1]-t[1])**2+(ActionValues[2]-t[2])**2+(ActionValues[3]-t[3])**2
    +(ActionValues[4]-t[4])**2+(ActionValues[5]-t[5])**2+(ActionValues[6]-t[6])**2+(ActionValues[7]-t[7])**2
    +(ActionValues[8]-t[8])**2+(ActionValues[9]-t[9])**2+(ActionValues[10]-t[10])**2+(ActionValues[11]-t[11])**2
    +(ActionValues[12]-t[12])**2+(ActionValues[3]-t[13])**2+(ActionValues[14]-t[14])**2+(ActionValues[15]-t[15])**2
    +(ActionValues[16]-t[16])**2+(ActionValues[17]-t[17])**2+(ActionValues[18]-t[18])**2+(ActionValues[19]-t[19])**2
    +(ActionValues[20]-t[20])**2+(ActionValues[21]-t[21])**2+(ActionValues[22]-t[22])**2+(ActionValues[23]-t[23])**2
    +(ActionValues[24]-t[24])**2+(ActionValues[25]-t[25])**2+(ActionValues[26]-t[26])**2+(ActionValues[27]-t[27])**2
    +(ActionValues[28]-t[28])**2+(ActionValues[29]-t[29])**2+(ActionValues[30]-t[30])**2+(ActionValues[31]-t[31])**2
    +(ActionValues[32]-t[32])**2+(ActionValues[33]-t[33])**2+(ActionValues[34]-t[34])**2+(ActionValues[35]-t[35])**2
    ))





      #EXPLORATION ACTION
            if random.uniform(0,1)>1-epsilon:# exploration condition
                #EXPLORATION ACTION
            #if random.uniform(0,1)>1-epsilon:# exploration condition
                A=random.choice(Actions)     #selecting a random action 
    #STATE (1,1)##########################################################################

                if S==(1,1):
                    if A=='U':
                        ActionValues[0]+=alpha*(r+gamma*max(ActionValues[4:8])-ActionValues[0]) #Q-learning update rule
                        S=(1,2)

                    elif A=='D':
                        ActionValues[1]+=alpha*(r+gamma*max(ActionValues[0:4])-ActionValues[1])

                        S=(1,1)

                    elif A=='L':
                        ActionValues[2]+=alpha*(r+gamma*max(ActionValues[0:4])-ActionValues[2])
                        S=(1,1)

                    elif A=='R':
                        ActionValues[3]+=alpha*(r+gamma*max(ActionValues[16:20])-ActionValues[3])
                        S=(2,1)

                #STATE (1,2)###############################################################

                elif S==(1,2):
                    if A=='U':
                        ActionValues[4]+=alpha*(r+gamma*max(ActionValues[8:12])-ActionValues[4])
                        S=(1,3)

                    elif A=='D':
                        ActionValues[5]+=alpha*(r+gamma*max(ActionValues[0:4])-ActionValues[5])
                        S=(1,1)

                    elif A=='L':
                        ActionValues[6]+=alpha*(r+gamma*max(ActionValues[4:8])-ActionValues[6])
                        S=(1,2)

                    elif A=='R':
                        ActionValues[7]+=alpha*(r+gamma*max(ActionValues[4:8])-ActionValues[7])
                        S=(2,1)

                #STATE (1,3)#################################################################

                elif S==(1,3):
                    if A=='U':
                        ActionValues[8]+=alpha*(r+gamma*max(ActionValues[8:12])-ActionValues[8])
                        S=(1,3)

                    elif A=='D':
                        ActionValues[9]+=alpha*(r+gamma*max(ActionValues[4:8])-ActionValues[9])
                        S=(1,2)

                    elif A=='L':
                        ActionValues[10]+=alpha*(r+gamma*max(ActionValues[8:12])-ActionValues[10])
                        S=(1,3)

                    elif A=='R':
                        ActionValues[11]+=alpha*(r+gamma*max(ActionValues[12:16])-ActionValues[11])
                        S=(2,3)

                #STATE (2,1)##################################################################
                elif S==(2,1):
                    if A=='U':
                        ActionValues[16]+=alpha*(r+gamma*max(ActionValues[16:20])-ActionValues[16])
                        S=(2,1)

                    elif A=='D':
                        ActionValues[17]+=alpha*(r+gamma*max(ActionValues[16:20])-ActionValues[17])
                        S=(2,1)

                    elif A=='L':
                        ActionValues[18]+=alpha*(r+gamma*max(ActionValues[0:4])-ActionValues[18])
                        S=(1,1)

                    elif A=='R':
                        ActionValues[19]+=alpha*(r+gamma*max(ActionValues[20:24])-ActionValues[19])
                        S=(3,1)

                #STATE (2,3)#####################################################################
                elif S==(2,3):
                    if A=='U':
                        ActionValues[12]+=alpha*(r+gamma*max(ActionValues[12:16])-ActionValues[12])
                        S=(2,3)

                    elif A=='D':
                        ActionValues[13]+=alpha*(r+gamma*max(ActionValues[12:16])-ActionValues[13])
                        S=(2,3)

                    elif A=='L':
                        ActionValues[14]+=alpha*(r+gamma*max(ActionValues[8:12])-ActionValues[14])
                        S=(1,3)

                    elif A=='R':
                        ActionValues[15]+=alpha*(r+gamma*max(ActionValues[28:32])-ActionValues[15])
                        S=(3,3)

                #STATE (3,1)#######################################################################


                elif S==(3,1):
                    if A=='U':
                        ActionValues[20]+=alpha*(r+gamma*max(ActionValues[24:28])-ActionValues[20])
                        S=(3,2)

                    elif A=='D':
                        ActionValues[21]+=alpha*(r+gamma*max(ActionValues[20:24])-ActionValues[21])
                        S=(2,3)

                    elif A=='L':
                        ActionValues[22]+=alpha*(r+gamma*max(ActionValues[16:20])-ActionValues[22])
                        S=(2,1)

                    elif A=='R':
                        ActionValues[23]+=alpha*(r+gamma*max(ActionValues[32:])-ActionValues[23])
                        S=(4,1)

                #STATE (3,2)#####################################################################


                elif S==(3,2):
                    if A=='U':
                        ActionValues[24]+=alpha*(r+gamma*max(ActionValues[28:32])-ActionValues[24])
                        S=(3,2)

                    elif A=='D':
                        ActionValues[25]+=alpha*(r+gamma*max(ActionValues[20:24])-ActionValues[25])

                        S=(2,1)

                    elif A=='L':
                        ActionValues[26]+=alpha*(r+gamma*max(ActionValues[24:28])-ActionValues[26])

                        S=(3,2)

                    elif A=='R':
                        ActionValues[27]+=alpha*(Rr+gamma*max(0,0,0,0)-ActionValues[27])

                        S=(4,2)

                #STATE (3,3)########################################################################


                elif S==(3,3):
                    if A=='U':
                        ActionValues[28]+=alpha*(r+gamma*max(ActionValues[28:32])-ActionValues[28])

                        S=(3,3)

                    elif A=='D':
                        ActionValues[29]+=alpha*(r+gamma*max(ActionValues[24:28])-ActionValues[29])

                        S=(3,2)

                    elif A=='L':
                        ActionValues[30]+=alpha*(r+gamma*max(ActionValues[12:16])-ActionValues[30])

                        S=(2,3)

                    elif A=='R':
                        ActionValues[31]+=alpha*(R+gamma*max(0,0,0,0)-ActionValues[31])
                        S=(4,3)

                #STATE (4,1)#######################################################################
                elif S==(4,1):
                    if A=='U':
                        ActionValues[32]+=alpha*(Rr+gamma*max(0,0,0,0)-ActionValues[32])
                        S=(4,2)

                    elif A=='D':
                        ActionValues[33]+=alpha*(r+gamma*max(ActionValues[32:])-ActionValues[33])
                        S=(4,1)

                    elif A=='L':
                        ActionValues[34]+=alpha*(r+gamma*max(ActionValues[20:24])-ActionValues[34])

                        S=(3,1)

                    elif A=='R':
                        ActionValues[35]+=alpha*(r+gamma*max(ActionValues[32:])-ActionValues[35])

                        S=(4,1)

                        
                        
            if random.uniform(0,1)<1-epsilon: #condition for exploitation

                #STATE(1,1)
                if S==(1,1):
                    c=ActionValues.index(max(ActionValues[:4])) #selecting the maximum action in state (1,1)
                    if c==0:#up
                        ActionValues[c]+=alpha*(r+gamma*max(ActionValues[4:8])-ActionValues[c])
                        S=(1,2)

                    if c==1:#down
                        ActionValues[c]+=alpha*(r+gamma*max(ActionValues[0:4])-ActionValues[c])
                        S=(1,1)

                    if c==2:#left
                        ActionValues[c]+=alpha*(r+gamma*max(ActionValues[0:4])-ActionValues[c])
                        S=(1,1)

                    if c==3:#right
                        ActionValues[c]+=alpha*(r+gamma*max(ActionValues[16:20])-ActionValues[c])
                        S=(1,1)

                #STATE(1,2)
                if S==(1,2):
                    c=ActionValues.index(max(ActionValues[4:8]))
                    if c==4:#up
                        ActionValues[c]+=alpha*(r+gamma*max(ActionValues[8:12])-ActionValues[c])
                        S=(1,3)

                    if c==5:#down
                        ActionValues[c]+=alpha*(r+gamma*max(ActionValues[0:4])-ActionValues[c])
                        S=(1,1)

                    if c==6:#left
                        ActionValues[c]+=alpha*(r+gamma*max(ActionValues[4:8])-ActionValues[c])
                        S=(1,2)

                    if c==7:#right
                        ActionValues[c]+=alpha*(r+gamma*max(ActionValues[4:8])-ActionValues[c])
                        S=(1,2)

                #STATE(1,3)
                if S==(1,3):
                    c=ActionValues.index(max(ActionValues[8:12]))
                    if c==8:#up
                        ActionValues[c]+=alpha*(r+gamma*max(ActionValues[8:12])-ActionValues[c])
                        S=(1,3)

                    if c==5:#down
                        ActionValues[c]+=alpha*(r+gamma*max(ActionValues[4:8])-ActionValues[c])
                        S=(1,2)

                    if c==6:#left
                        ActionValues[c]+=alpha*(r+gamma*max(ActionValues[8:12])-ActionValues[c])
                        S=(1,3)

                    if c==7:#right
                        ActionValues[c]+=alpha*(r+gamma*max(ActionValues[12:16])-ActionValues[c])
                        S=(2,3)

                #STATE(2,3)
                if S==(2,3):
                    c=ActionValues.index(max(ActionValues[12:16]))
                    if c==12:#up
                        ActionValues[c]+=alpha*(r+gamma*max(ActionValues[12:16])-ActionValues[c])
                        S=(2,3)

                    if c==13:#down
                        ActionValues[c]+=alpha*(r+gamma*max(ActionValues[12:16])-ActionValues[c])
                        S=(2,3)

                    if c==14:#left
                        ActionValues[c]+=alpha*(r+gamma*max(ActionValues[8:12])-ActionValues[c])
                        S=(1,3)

                    if c==15:#right
                        ActionValues[c]+=alpha*(r+gamma*max(ActionValues[28:32])-ActionValues[c])
                        S=(3,3)

                #STATE(2,1)
                if S==(2,1):
                    c=ActionValues.index(max(ActionValues[16:20]))
                    if c==16:#up
                        ActionValues[c]+=alpha*(r+gamma*max(ActionValues[16:20])-ActionValues[c])
                        S=(2,1)

                    if c==17:#down
                        ActionValues[c]+=alpha*(r+gamma*max(ActionValues[16:20])-ActionValues[c])
                        S=(2,1)

                    if c==18:#left
                        ActionValues[c]+=alpha*(r+gamma*max(ActionValues[:4])-ActionValues[c])
                        S=(1,1)

                    if c==19:#right
                        ActionValues[c]+=alpha*(r+gamma*max(ActionValues[20:24])-ActionValues[c])
                        S=(3,1)


                #STATE(3,1)
                if S==(3,1):
                    c=ActionValues.index(max(ActionValues[20:24]))
                    if c==20:#up
                        ActionValues[c]+=alpha*(r+gamma*max(ActionValues[24:28])-ActionValues[c])
                        S=(3,2)

                    if c==21:#down
                        ActionValues[c]+=alpha*(r+gamma*max(ActionValues[20:24])-ActionValues[c])
                        S=(3,1)

                    if c==22:#left
                        ActionValues[c]+=alpha*(r+gamma*max(ActionValues[16:20])-ActionValues[c])
                        S=(2,1)

                    if c==22:#right
                        ActionValues[c]+=alpha*(r+gamma*max(ActionValues[20:24])-ActionValues[c])
                        S=(4,1)


                #STATE(3,2)
                if S==(3,2):
                    c=ActionValues.index(max(ActionValues[24:28]))
                    if c==24:#up
                        ActionValues[c]+=alpha*(r+gamma*max(ActionValues[28:32])-ActionValues[c])
                        S=(3,3)

                    if c==25:#down
                        ActionValues[c]+=alpha*(r+gamma*max(ActionValues[20:24])-ActionValues[c])
                        S=(3,1)

                    if c==26:#left
                        ActionValues[c]+=alpha*(r+gamma*max(ActionValues[24:28])-ActionValues[c])
                        S=(3,2)

                    if c==27:#right
                        # the q values of terminal states are all 0's
                        ActionValues[c]+=alpha*(Rr+gamma*max(0,0,0,0)-ActionValues[c])
                        S=(4,2)

                #STATE(3,3)
                if S==(3,3):
                    c=ActionValues.index(max(ActionValues[28:32]))
                    if c==28:#up
                        ActionValues[c]+=alpha*(r+gamma*max(ActionValues[28:32])-ActionValues[c])
                        S=(3,3)

                    if c==29:#down
                        ActionValues[c]+=alpha*(r+gamma*max(ActionValues[24:28])-ActionValues[c])
                        S=(3,2)

                    if c==30:#left
                        ActionValues[c]+=alpha*(r+gamma*max(ActionValues[12:16])-ActionValues[c])
                        S=(2,3)

                    if c==31:#right
                        ActionValues[c]+=alpha*(R+gamma*max(0,0,0,0)-ActionValues[c])
                        S=(4,1)


                #STATE(4,1)
                if S==(4,1):
                    c=ActionValues.index(max(ActionValues[32:]))
                    if c==32:#up
                        ActionValues[c]+=alpha*(Rr+gamma*max(0,0,0,0)-ActionValues[c])
                        S=(4,2)

                    if c==33:#down
                        ActionValues[c]+=alpha*(r+gamma*max(ActionValues[32:])-ActionValues[c])
                        S=(4,1)

                    if c==34:#left
                        ActionValues[c]+=alpha*(r+gamma*max(ActionValues[20:24])-ActionValues[c])
                        S=(3,1)

                    if c==35:#right
                        ActionValues[c]+=alpha*(r+gamma*max(ActionValues[32:])-ActionValues[c])
                        S=(4,1)
        

    time.sleep(1)
    end = time.time()
    print(f"Runtime of the program is {end - start}")
    return sum_mse,end-start
    
```
